```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("internship_train.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data_x = data.drop('target', axis = 1)
```

```python
import matplotlib.pyplot as plt
plt.plot(data_x.mean())
```

```python
x1,x2,y1,y2 =  train_test_split(data_x, data['target'], random_state=0, train_size=0.5)
```

```python
forest = RandomForestRegressor(100)
```

```python
forest.fit(x1, y1)
```

```python
y_predict = forest.predict(x2)
print(y2.iloc[:5], y_predict[:5])
```

```python
print("accuracy")
y2/y_predict*100
```
